```python
df['Time'] = pd.to_datetime(df['Time_1Hz'])
```

```python
def align_time_series(df, freq='1S'):
    df.set_index('Time', inplace=True)
    resampled_df = df.resample(freq).mean().interpolate()
    return resampled_df.reset_index()
```
